## 1) Import Packages

```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

from itertools import permutations

from sklearn.model_selection import validation_curve, learning_curve, train_test_split

from sklearn import preprocessing
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder, normalize
from sklearn.metrics import f1_score, silhouette_score, homogeneity_score, accuracy_score
from sklearn.model_selection import ShuffleSplit
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV, learning_curve



from sklearn.cluster import KMeans, FeatureAgglomeration
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA, FactorAnalysis, FastICA
from time import time
from sklearn.metrics import mean_squared_error, rand_score, adjusted_rand_score, adjusted_mutual_info_score
from sklearn.random_projection import GaussianRandomProjection 
from sklearn.decomposition import TruncatedSVD

random_seed = 123


import warnings
warnings.filterwarnings('ignore')
```

## 2) Load Data

#### Load Churn Data

```python
with open("./Data/churn.csv") as f:
    f.readline()  # skip the header
    data_reader = csv.reader(f, delimiter=",")
    X = []
    y = []
    for row in data_reader:
        features = row[1:-1]
        label = row[-1]
        X.append([float(feature) for feature in features])
        y.append(int(label))
        
enc = OrdinalEncoder().fit(X)
X = enc.transform(X)
X = normalize(X)
le = preprocessing.LabelEncoder().fit(y)
y = le.transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size=0.2)
```

## 4) Helper Functions

## 5) Dimensionality Reduction: PCA

```python
num_comps = range(len(X_train[0]))
explained_var = []
reconstruction_err = []
for dim in num_comps:
    pca = PCA(n_components=dim, random_state=random_seed)
    pca.fit_transform(X_train)
    
    # reconstruction
    x_prime = pca.fit_transform(X_train)
    A = np.linalg.pinv(pca.components_.T)
    reconstructed = np.dot(x_prime, A)
    rc_err = mean_squared_error(X_train, reconstructed)    
    explained_var.append(sum(pca.explained_variance_ratio_))
    reconstruction_err.append(rc_err)
```

```python
num_comps = len(X_train[0])
comps_range = range(len(X_train[0]))

pca = PCA(n_components=num_comps, random_state=random_seed)
pca.fit_transform(X_train)

plt.plot(comps_range, explained_var)
plt.title('PCA Explained Variance vs. Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.show()
```

Determine number of components required for a >= 95% explained variance

```python
for i in range(num_comps):
    if explained_var[i] >= 0.95:
        print('Minimum number of dimensions required for >= 95% explained Variance: {}'.format(i))
        break
```

source: https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html <br>
If 0 < n_components < 1 and svd_solver == 'full', select the number of components such that the amount of variance that needs to be explained is greater than the percentage specified by n_components.

```python
# set number of components to 95% explained variance threshold
pca = PCA(0.95)
pca.fit(X_train)
pca_train = pca.transform(X_train)
pca_test = pca.transform(X_test)
```

```python
clf = MLPClassifier(hidden_layer_sizes=(100,),learning_rate_init=0.05, random_state=10)
start = time()
clf.fit(pca_train, y_train)
end = time()
print("training time: {}".format(end-start))
print("train score: ", clf.score(pca_train, y_train))
print("test score: ", clf.score(pca_test, y_test))
```

# 6) Dimensionality Reduction: ICA

### 6.1.a Reduce Dimensinality with ICA

```python
num_comps = len(X_train[0])
comps_range = range(2, num_comps, 2)
reconstruction_err = []
ica_times = []
for dim in comps_range:
    ica = FastICA(n_components=dim, random_state=random_seed)
    # reconstruction
    x_prime = ica.fit_transform(X_train)
    A = np.linalg.pinv(ica.components_.T)
    reconstructed = np.dot(x_prime, A)
    rc_err = mean_squared_error(X_train, reconstructed)    
    reconstruction_err.append(rc_err)
```

```python
plt.plot(comps_range, reconstruction_err)
plt.xlabel('Number of Components')
plt.ylabel('Reconstruction Error')
plt.title('ICA: Reconstruction Error (Churn Dataset)')
plt.show()
```

#### Set apply optimal ICA Dimensionality Reduction

```python
ica = FastICA(n_components=4)
ica.fit(X_train)
ica_train = ica.transform(X_train)
ica_test = ica.transform(X_test)
```

### 6.2.a Apply ICA reduced data to Neural Network
- Same performance as PCA reduced data, but trained 35% faster

### Make Learning Curve

```python
def NNGridSearch(X_train, y_train):

    hl_range = [5,20,40,60,80,100]
    lr_range = [0.01,0.05,0.1]
    param_grid = {'hidden_layer_sizes': hl_range, 'learning_rate_init': lr_range}

    net = GridSearchCV(estimator = MLPClassifier(random_state=0),
                       param_grid=param_grid, cv=5)
    net.fit(X_train, y_train)
    print("Best NN HypterParamters")
    print(net.best_params_)
    return net.best_params_['hidden_layer_sizes'], net.best_params_['learning_rate_init']


def plot_hidden_layers(X_train, y_train, X_test, y_test):

    train_score = []
    test_score = []
    num_hl = [5,20,40,60,80,100]
    for hl in num_hl:         
        clf = MLPClassifier(hidden_layer_sizes=(hl,), random_state=10)
        clf.fit(X_train, y_train)
        test_pred = clf.predict(X_test)
        train_pred = clf.predict(X_train)
        test_score.append(f1_score(y_test, test_pred))
        train_score.append(f1_score(y_train, train_pred))
        
    
    plt.plot(num_hl, test_score, 'o-', color='g', label='Test Score')
    plt.plot(num_hl, train_score, 'o-', color = 'r', label='Training Score')
    plt.ylabel('Model F1 Score')
    plt.xlabel('Number of Hidden Layers')
    
    plt.title("Number of hidden layers vs. F1 Score")
    plt.legend(loc='best')
    plt.tight_layout()
    plt.show()
    
def plot_learning_rate(X_train, y_train, X_test, y_test):

    train_score = []
    test_score = []
    lr_range = [0.01,0.05,0.1]
    for lr in lr_range:         
        clf = MLPClassifier(learning_rate_init=lr, random_state=10)
        clf.fit(X_train, y_train)
        test_pred = clf.predict(X_test)
        train_pred = clf.predict(X_train)
        test_score.append(f1_score(y_test, test_pred))
        train_score.append(f1_score(y_train, train_pred))
        
    
    plt.plot(lr_range, test_score, 'o-', color='g', label='Test Score')
    plt.plot(lr_range, train_score, 'o-', color = 'r', label='Training Score')
    plt.ylabel('Model F1 Score')
    plt.xlabel('Learning Rate Init')
    
    plt.title("Learning Rate Init vs. F1 Score")
    plt.legend(loc='best')
    plt.tight_layout()
    plt.show()
```

```python
best_num_hl, best_lr = NNGridSearch(pca_train, y_train)
best_nn = MLPClassifier(hidden_layer_sizes=(best_num_hl,),learning_rate_init=best_lr, random_state=10)
```

```python
def plot_learning_curve(x_new, name):

    fig, axes = plt.subplots(1, 1, figsize=(6.4, 4.8))
    axes.set_title("{} Learning Curve".format(name))
    axes.set_xlabel("Training examples")
    axes.set_ylabel("Score")

    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=10)
    pca_train_sizes, pca_train_scores, pca_test_scores, pca_fit_times, pca_score_times = learning_curve(
        best_nn,
        x_new,
        y_train,
        train_sizes=np.linspace(0.1, 1.0, 5),
        cv=cv,
        return_times=True
    )
    train_sizes, train_scores, test_scores, fit_times, score_times = learning_curve(
        best_nn,
        X,
        y,
        train_sizes=np.linspace(0.1, 1.0, 5),
        cv=cv,
        return_times=True
    )

    train_scores_mean = np.mean(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    
    pca_fit_times_mean = np.mean(pca_fit_times, axis=1)
    fit_times_mean = np.mean(fit_times, axis=1)
    
    pca_score_times_mean = np.mean(pca_score_times, axis=1)
    score_times_mean = np.mean(score_times, axis=1)
    
    pca_train_scores_mean = np.mean(pca_train_scores, axis=1)
    pca_test_scores_mean = np.mean(pca_test_scores, axis=1)

    # Plot learning curve
    axes.grid()
    axes.plot(
        train_sizes, train_scores_mean, "o-", color="r", label="No DR: Training score"
    )
    axes.plot(
        train_sizes, test_scores_mean, "o-", color="g", label="No DR: Cross-Validation score"
    )
    axes.plot(
        pca_train_sizes, pca_train_scores_mean, "o-", color="k", label="{}: Training score".format(name)
    )
    axes.plot(
        pca_train_sizes, pca_test_scores_mean, "o-", color="b", label="{}: Cross-Validation score".format(name)
    )
    axes.legend(loc="best")
    return fig, train_sizes, pca_fit_times_mean, fit_times_mean, pca_score_times_mean, score_times_mean
```
