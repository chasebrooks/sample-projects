## 1) Import Packages

```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

from itertools import permutations

from sklearn.model_selection import validation_curve, learning_curve, train_test_split

from sklearn import preprocessing
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder, normalize
from sklearn.metrics import f1_score, silhouette_score, homogeneity_score, accuracy_score
from sklearn.model_selection import ShuffleSplit


from sklearn.cluster import KMeans, FeatureAgglomeration
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA, FactorAnalysis, FastICA
from time import time
from sklearn.metrics import mean_squared_error, rand_score, adjusted_rand_score, adjusted_mutual_info_score
from sklearn.random_projection import GaussianRandomProjection 
from sklearn.decomposition import TruncatedSVD

from scipy.stats import kurtosis


random_seed = 123


import warnings
warnings.filterwarnings('ignore')
```

## 2) Load Data

#### Load Human Activity Data

```python
with open("./Data/human_activity.csv") as f:
    f.readline()  # skip the header
    data_reader = csv.reader(f, delimiter=",")
    X = []
    y = []
    for row in data_reader:
        features = row[1:-1]
        label = row[-1]
        X.append([float(feature) for feature in features])
        y.append(int(label))
        
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size=0.1)

X_train = np.array(X_train)
y_train = np.array(y_train)
X_test = np.array(X_test)
y_test = np.array(y_test)
```

### Helper Functions

```python
def analyze_kmeans(X, y, title):

    k_range = [1] + [i for i in range(5,56,5)]
    inertia = []
    rand_score_lst = []
    mutual_info_score = []
    kmeans_times = []
    for k in k_range:
        start = time()
        kmeans = KMeans(n_clusters=k, random_state=random_seed)
        kmeans.fit(X)
        end = time()
        preds = kmeans.labels_
        rand_score_lst.append(rand_score(preds, y))
        mutual_info_score.append(adjusted_mutual_info_score(preds, y))
        kmeans_times.append(end-start)
        inertia.append(kmeans.inertia_)
        
    plt.clf()
        
    fig, (ax1, ax2) = plt.subplots(2, sharex=True)
    fig.subplots_adjust(hspace=.6)
    
    fig.suptitle(title)
    
    # Plot intertia vs k
    ax1.plot(k_range, inertia, label='Inertia', color='tab:green')
    ax1.set_ylabel('Inertia') 
    ax1.set_title('Intra-cluster Sum of Squared Errors', fontdict = {'fontsize' : 10})
    ax1.legend()
    
    # Plot rand score vs k
    ax2.plot(k_range, rand_score_lst, label='Rand Score', color='tab:blue')
    ax2.plot(k_range, mutual_info_score, label='Mutual Info Score', color='tab:orange')
    ax2.set_xlabel('K')
    ax2.set_title('Cluster Performance', fontdict = {'fontsize' : 10})
    ax2.legend()
    
    return fig, kmeans_times

def analyze_gmm(X, y, title):

    comps_range = [1] + [i for i in range(5,56,5)]
    bic = []
    rand_score_lst = []
    mutual_info_score = []
    gmm_times = []
    for n_components in comps_range:
        start = time()
        gmm = GaussianMixture(n_components=n_components, random_state=random_seed)
        gmm.fit(X)
        end = time()
        preds = gmm.predict(X)
        rand_score_lst.append(rand_score(preds, y))
        mutual_info_score.append(adjusted_mutual_info_score(preds, y))
        bic.append(gmm.bic(X))
        gmm_times.append(end-start)
        
    plt.clf()
        
    fig, (ax1, ax2) = plt.subplots(2, sharex=True)
    fig.subplots_adjust(hspace=.6)
    
    fig.suptitle(title)
    
    # Plot intertia vs k
    ax1.plot(comps_range, bic, label='BIC', color='tab:green')
    ax1.set_ylabel('BIC') 
    ax1.set_title('Bayesian Information Coefficient', fontdict = {'fontsize' : 10})
    ax1.legend()
    
    # Plot rand score vs num components
    ax2.plot(comps_range, rand_score_lst, label='Rand Score', color='tab:blue')
    ax2.plot(comps_range, mutual_info_score, label='Mutual Info Score', color='tab:orange')
    ax2.set_xlabel('Num Components')
    ax2.set_title('GMM Performance', fontdict = {'fontsize' : 10})
    ax2.legend()
    
    return fig, gmm_times

def plot_training_times(kmeans_times, gmm_times, title):
    x_range = [1] + [i for i in range(5,56,5)]
    plt.clf()
    fig = plt.figure()
    plt.plot(x_range, gmm_times, label='GMM')
    plt.plot(x_range, kmeans_times, label='KMeans')
    plt.xlabel('Number of components')
    plt.ylabel('Time (s)')
    plt.title(title)
    plt.legend()
    return fig
```

## 3) Functions to graph outputs

### 4.1) K-Means Clustering

#### 4.1.a Find Optimal K using elbow method
Source: https://www.geeksforgeeks.org/elbow-method-for-optimal-value-of-k-in-kmeans/ <br>
Inertia: It is the sum of squared distances of samples to their closest cluster center. <br>
To determine the optimal number of clusters, we have to select the value of k at the “elbow” ie the point after which the distortion/inertia start decreasing in a linear fashion. 

```python
fig, kmeans_times_full = analyze_kmeans(X_train, y_train, 'Human Activity training set')
plt.show()
```

#### Determine Optimal K Value using Silhouette Analysis
- Source: https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html <br>
- Silhouette coefficients (as these values are referred to as) near +1 indicate that the sample is far away from the neighboring clusters. A value of 0 indicates that the sample is on or very close to the decision boundary between two neighboring clusters and negative values indicate that those samples might have been assigned to the wrong cluster.

#### 4.1.b Determine Cluster Performance of KMeans with k=10

```python
clf = KMeans(n_clusters=10, random_state=random_seed, max_iter=300)
clf.fit(X_test)
preds = clf.labels_

print('Rand Score: {}'.format(rand_score(preds, y_test)))
print('Adjusted Rand Score: {}'.format(adjusted_rand_score(preds, y_test)))
print('Adjusted Mutual Info Score: {}'.format(adjusted_mutual_info_score(preds, y_test)))
```

## 5) Dimensionality Reduction: PCA

```python
comps_range = [1] + [i for i in range(5,104,5)]
explained_var = []
reconstruction_err = []
for dim in comps_range:
    pca = PCA(n_components=dim, random_state=random_seed)
    pca_data = pca.fit_transform(X_train)
    
    # reconstruction
    x_prime = pca.fit_transform(X_train)
    A = np.linalg.pinv(pca.components_.T)
    reconstructed = np.dot(x_prime, A)
    rc_err = mean_squared_error(X_train, reconstructed)    
    explained_var.append(sum(pca.explained_variance_ratio_))
    reconstruction_err.append(rc_err)
```

```python
plt.clf()
        
fig, (ax1, ax2) = plt.subplots(2, sharex=True)
fig.subplots_adjust(hspace=.6)

comps_range = [1] + [i for i in range(5,104,5)]

pca = PCA(n_components=100, random_state=random_seed)
pca.fit_transform(X_train)
eigenvalues = pca.singular_values_
eigenvalues = [eigenvalues[i-1] for i in comps_range]

ax1.plot(comps_range, explained_var, color='tab:orange')
ax1.set_title('PCA Explained Variance (Human Activity Dataset)')
ax1.set_xlabel('Number of Components')
ax1.set_ylabel('Explained Variance')

ax2.plot(comps_range, eigenvalues, color='tab:blue')
ax2.set_title('Eigenvalue Distribution (Human Activity Dataset)')
ax2.set_xlabel('Number of Components')
ax2.fill_between(comps_range, eigenvalues)


plt.show()
```

Determine number of components required for a >= 95% explained variance

```python
for i in comps_range:
    if explained_var[i] >= 0.95:
        print('Minimum number of dimensions required for >= 95% explained Variance: {}'.format(i))
        break
```

# 6) Dimensionality Reduction: ICA

### 6.1.a Reduce Dimensinality with ICA

```python
num_comps = len(X_train[0])
comps_range = [1] + [i for i in range(5,104,5)]
reconstruction_err = []
kurtosis = []
ica_times = []
for dim in comps_range:
        
    ica = FastICA(n_components=dim, random_state=random_seed)
    
    # reconstruction
    x_prime = ica.fit_transform(X_train)
    kurt_df = pd.DataFrame(x_prime)
    kurt_df = kurt_df.kurt(axis=0)
    kurtosis.append(kurt_df.abs().mean())

    A = np.linalg.pinv(ica.components_.T)
    reconstructed = np.dot(x_prime, A)
    rc_err = mean_squared_error(X_train, reconstructed)    
    reconstruction_err.append(rc_err)
```

```python
plt.clf()
        
fig, (ax1, ax2) = plt.subplots(2, sharex=True)
fig.subplots_adjust(hspace=.6)

comps_range = [1] + [i for i in range(5,104,5)]


ax1.plot(comps_range, reconstruction_err, color='tab:orange')
ax1.set_xlabel('Number of Components')
ax1.set_ylabel('Reconstruction Error')
ax1.set_title('ICA: Reconstruction Error (Human Activity Dataset)')

ax2.plot(comps_range, kurtosis, color='tab:blue')
ax2.set_xlabel('Number of Components')
ax2.set_ylabel('Kurtosis')
ax2.set_title('ICA: Kurtosis (Human Activity Dataset)')
ax2.set_ylim(0, 150)
ax2.fill_between(comps_range, kurtosis)


plt.show()
```

#### Set apply optimal ICA Dimensionality Reduction

```python
ica = FastICA(n_components=15)
ica.fit(X_train)
ica_train = ica.transform(X_train)
ica_test = ica.transform(X_test)
```

## 8) TruncatedSVD


### 8.1 Dimenionality Reduction with Feature Agglomeration

```python
comps_range = [1] + [i for i in range(5,104,5)]
reconstruction_err = []
tsvd_times = []
for dim in comps_range:
    tsvd = TruncatedSVD(n_components=dim, random_state=random_seed)
    # reconstruction
    x_prime = tsvd.fit_transform(X_train)
    A = np.linalg.pinv(tsvd.components_.T)
    reconstructed = np.dot(x_prime, A)
    rc_err = mean_squared_error(X_train, reconstructed)    
    reconstruction_err.append(rc_err)
```

```python
# Plot reconstruction error vs num components to find best num of components
plt.plot(comps_range, reconstruction_err)
plt.xlabel('Number of Components')
plt.ylabel('Reconstruction Error')
plt.title('TruncatedSVD: Reconstruction Error (Human Activity Dataset)')
plt.show()
```
