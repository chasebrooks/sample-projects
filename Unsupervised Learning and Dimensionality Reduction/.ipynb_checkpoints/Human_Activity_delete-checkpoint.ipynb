## 1) Import Packages

```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

from itertools import permutations

from sklearn.model_selection import validation_curve, learning_curve, train_test_split

from sklearn import preprocessing
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder, normalize
from sklearn.metrics import f1_score, silhouette_score, homogeneity_score, accuracy_score
from sklearn.model_selection import ShuffleSplit


from sklearn.cluster import KMeans, FeatureAgglomeration
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA, FactorAnalysis, FastICA
from time import time
from sklearn.metrics import mean_squared_error, rand_score, adjusted_rand_score, adjusted_mutual_info_score
from sklearn.random_projection import GaussianRandomProjection 
from sklearn.decomposition import TruncatedSVD

random_seed = 123


import warnings
warnings.filterwarnings('ignore')
```

## 2) Load Data

#### Load Human Activity Data

```python
with open("./Data/human_activity.csv") as f:
    f.readline()  # skip the header
    data_reader = csv.reader(f, delimiter=",")
    X = []
    y = []
    for row in data_reader:
        features = row[1:-1]
        label = row[-1]
        X.append([float(feature) for feature in features])
        y.append(int(label))
        
# enc = OrdinalEncoder().fit(X)
# X = enc.transform(X)
# X = normalize(X)
# le = preprocessing.LabelEncoder().fit(y)
# y = le.transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, test_size=0.1)

X_train = np.array(X_train)
y_train = np.array(y_train)
X_test = np.array(X_test)
y_test = np.array(y_test)
```

## 3) Functions to graph outputs

### 4.1) K-Means Clustering

#### 4.1.a Find Optimal K using elbow method
Source: https://www.geeksforgeeks.org/elbow-method-for-optimal-value-of-k-in-kmeans/ <br>
Inertia: It is the sum of squared distances of samples to their closest cluster center. <br>
To determine the optimal number of clusters, we have to select the value of k at the “elbow” ie the point after which the distortion/inertia start decreasing in a linear fashion. 

```python
k_range = [1] + [i for i in range(2,52,2)]
inertia = []
kmeans_times_full = []
for i, k in enumerate(k_range):
    start = time()
    kmeans = KMeans(n_clusters=k, random_state=random_seed)
    kmeans.fit(X_train)
    end = time()
    kmeans_times_full.append(end-start)
    inertia.append(kmeans.inertia_)
```

```python
# Plot intertia vs k to find best k
plt.plot(k_range, inertia)
plt.xlabel('K')
plt.ylabel('Inertia')
plt.title('KMeans Interia (Human Activity Dataset)')
plt.show()
```

#### Determine Optimal K Value using Silhouette Analysis
- Source: https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html <br>
- Silhouette coefficients (as these values are referred to as) near +1 indicate that the sample is far away from the neighboring clusters. A value of 0 indicates that the sample is on or very close to the decision boundary between two neighboring clusters and negative values indicate that those samples might have been assigned to the wrong cluster.

#### 4.1.b Determine Cluster Performance of KMeans with k=5

```python
clf = KMeans(n_clusters=5, random_state=random_seed, max_iter=300)
clf.fit(X_test)
preds = clf.labels_

print('Rand Score: {}'.format(rand_score(preds, y_test)))
print('Adjusted Rand Score: {}'.format(adjusted_rand_score(preds, y_test)))
print('Adjusted Mutual Info Score: {}'.format(adjusted_mutual_info_score(preds, y_test)))
```

## 4) Gaussian Mixture Model (Expectation Maximization)

- Source: https://scikit-learn.org/stable/auto_examples/mixture/plot_gmm_selection.html 
- Model selection can be performed with Gaussian Mixture Models using information-theoretic criteria (BIC)

```python
bic = []
components_range = [i for i in range(2,52,2)]
for n_components in components_range:
    gmm = GaussianMixture(n_components=n_components, random_state=random_seed)
    gmm.fit(X_train)
    bic.append(gmm.bic(X_train))
```

```python
# Plot bic vs k to find best k
plt.plot(components_range, bic)
plt.xlabel('Number of Components')
plt.ylabel('Bayes information criterion (BIC)')
plt.title('BIC (Human Activity Dataset)')
plt.show()
```

```python
clf = GaussianMixture(n_components=5, random_state=random_seed).fit(X_test)
preds = clf.predict(X_test)

print('Rand Score: {}'.format(rand_score(preds, y_test)))
print('Adjusted Rand Score: {}'.format(adjusted_rand_score(preds, y_test)))
print('Adjusted Mutual Info Score: {}'.format(adjusted_mutual_info_score(preds, y_test)))
```

## 5) Dimensionality Reduction: PCA

```python
num_comps = range(len(X_train[0]))
explained_var = []
reconstruction_err = []
for dim in num_comps:
    pca = PCA(n_components=dim, random_state=random_seed)
    pca.fit_transform(X_train)
    
    # reconstruction
    x_prime = pca.fit_transform(X_train)
    A = np.linalg.pinv(pca.components_.T)
    reconstructed = np.dot(x_prime, A)
    rc_err = mean_squared_error(X_train, reconstructed)    
    explained_var.append(sum(pca.explained_variance_ratio_))
    reconstruction_err.append(rc_err)
```

```python
num_comps = len(X_train[0])
comps_range = range(len(X_train[0]))

pca = PCA(n_components=num_comps, random_state=random_seed)
pca.fit_transform(X_train)

plt.plot(comps_range, explained_var)
plt.title('PCA Explained Variance vs. Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.show()
```

Determine number of components required for a >= 95% explained variance

```python
for i in range(num_comps):
    if explained_var[i] >= 0.95:
        print('Minimum number of dimensions required for >= 95% explained Variance: {}'.format(i))
        break
```

source: https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html <br>
If 0 < n_components < 1 and svd_solver == 'full', select the number of components such that the amount of variance that needs to be explained is greater than the percentage specified by n_components.

```python
# set number of components to 95% explained variance threshold
pca = PCA(0.95)
pca.fit(X_train)
pca_train = pca.transform(X_train)
pca_test = pca.transform(X_test)
```

```python
k_range = [1] + [i for i in range(2,52,2)]
inertia = []
kmeans_times_pca = []
for i, k in enumerate(k_range):
    start = time()
    kmeans = KMeans(n_clusters=k, random_state=random_seed)
    kmeans.fit(pca_train)
    end = time()
    kmeans_times_pca.append(end-start)
    inertia.append(kmeans.inertia_)
```

```python
# Plot intertia vs k to find best k
plt.plot(k_range, inertia)
plt.xlabel('K')
plt.ylabel('Inertia')
plt.title('PCA: KMeans Interia (Human Activity Dataset)')
plt.show()
```

```python
clf = KMeans(n_clusters=10, random_state=random_seed, max_iter=300)
clf.fit(pca_test)
preds = clf.labels_

print('Rand Score: {}'.format(rand_score(preds, y_test)))
print('Adjusted Rand Score: {}'.format(adjusted_rand_score(preds, y_test)))
print('Adjusted Mutual Info Score: {}'.format(adjusted_mutual_info_score(preds, y_test)))
```
