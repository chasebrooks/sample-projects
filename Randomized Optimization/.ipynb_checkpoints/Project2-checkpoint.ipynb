```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import mlrose_hiive as mlrose
from sklearn.metrics import f1_score, accuracy_score, classification_report
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
import time
```

## Toy Problem #1: Four Peaks

#### Initialize Problem data

```python
# returns min iteration idx of max fitness value
def find_max_idx(curve):
    curve = [i[0] for i in curve]
    max_value = max(curve)
    max_index = curve.index(max_value)
    return max_index, max_value
```

```python
fitness_sa = []
fitness_rhc = []
fitness_ga = []
fitness_mimic = []

iter_to_max_sa = []
iter_to_max_rhc = []
iter_to_max_ga = []
iter_to_max_mimic = []

time_sa = []
time_rhc = []
time_ga = []
time_mimic = []

## Plot effect of increasing problem size

range_values = [10,20,40,60,80]
np.random.seed(97)

for value in range_values:
    print("Starting Problem Size: {}".format(value) )
    fitness = mlrose.FourPeaks(t_pct = 0.1)
    problem = mlrose.DiscreteOpt(length = value, fitness_fn = fitness, maximize = True, max_val = 2)
    problem.set_mimic_fast_mode(True)
    init_state = np.random.randint(2, size = value)
    
    start = time.time()
    _, best_fitness_sa, best_curve_sa = mlrose.simulated_annealing(problem, curve = True)
    end = time.time()
    sa_time = end - start
    num_iter_to_max_sa, _ = find_max_idx(best_curve_sa)

    start = time.time()
    _, best_fitness_rhc, best_curve_rhc = mlrose.random_hill_climb(problem, curve = True)
    end = time.time()
    rhc_time = end - start
    num_iter_to_max_rhc, _ = find_max_idx(best_curve_rhc)

    start = time.time()
    _, best_fitness_ga, best_curve_ga = mlrose.genetic_alg(problem, curve = True)
    end = time.time()
    ga_time = end - start
    num_iter_to_max_ga, _ = find_max_idx(best_curve_ga)

    start = time.time()
    _, best_fitness_mimic, best_curve_mimic = mlrose.mimic(problem, curve = True)
    end = time.time()
    mimic_time = end - start
    num_iter_to_max_mimic, _ = find_max_idx(best_curve_mimic)

    fitness_sa.append(best_fitness_sa)
    fitness_rhc.append(best_fitness_rhc)
    fitness_ga.append(best_fitness_ga)
    fitness_mimic.append(best_fitness_mimic)
    
    iter_to_max_sa.append(num_iter_to_max_sa)
    iter_to_max_rhc.append(num_iter_to_max_rhc)
    iter_to_max_ga.append(num_iter_to_max_ga)
    iter_to_max_mimic.append(num_iter_to_max_mimic)
    
    time_sa.append(sa_time)
    time_rhc.append(rhc_time)
    time_ga.append(ga_time)
    time_mimic.append(mimic_time)


```

#### Problem Size vs. Fitness

```python
plt.figure()
plt.plot(range_values, fitness_sa, label = 'Simulated Annealing')
plt.plot(range_values, fitness_rhc, label = 'Randomized Hill Climb')
plt.plot(range_values, fitness_ga, label = 'Genetic Algorithm')
plt.plot(range_values, fitness_mimic, label = 'MIMIC')
plt.title('Fitness vs. Problem Size (Four Peaks)')
plt.xlabel('Problem Size')
plt.ylabel('Fitness')
plt.legend()
```

#### Plot Num Iterations to Max Fitness

```python
plt.figure()
plt.plot(range_values, iter_to_max_sa, label = 'Simulated Annealing')
plt.plot(range_values, iter_to_max_rhc, label = 'Randomized Hill Climb')
plt.plot(range_values, iter_to_max_ga, label = 'Genetic Algorithm')
plt.plot(range_values, iter_to_max_mimic, label = 'MIMIC')
plt.title('Number of Function Evalutations to Maximize Four Peaks')
plt.xlabel('Problem Size')
plt.ylabel('Number of iterations')
plt.legend()
```

#### Plot Training Time

```python
plt.figure()
plt.plot(range_values, time_sa, label = 'Simulated Annealing')
plt.plot(range_values, time_rhc, label = 'Randomized Hill Climb')
plt.plot(range_values, time_ga, label = 'Genetic Algorithm')
plt.plot(range_values, time_mimic, label = 'MIMIC')
plt.title('Wall Clock Time vs. Problem Size (Four Peaks)')
plt.legend()
plt.xlabel('Problem Size')
plt.ylabel('Computation Time (s)')
```

## Neural Network HyperParameter Tuning

#### Load Spam Data

```python
with open("./Data/spam.csv") as f:
    f.readline()  # skip the header
    data_reader = csv.reader(f, delimiter=",")
    X = []
    y = []
    for row in data_reader:
        features = row[1:-1]
        label = row[-1]
        X.append([float(feature) for feature in features])
        y.append(int(label))
```

#### Preprocess Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=97, test_size=0.2)
```

#### Tune weights with Gradient Descent (for comparison)

```python
clf = mlrose.NeuralNetwork(algorithm='gradient_descent', max_iters=1000, early_stopping=True)
clf.fit(X_train, y_train)
test_pred = clf.predict(X_test)
train_pred = clf.predict(X_train)
print(classification_report(y_test, test_pred))
```

#### Tune weights with Random Hill Climb

```python
clf = mlrose.NeuralNetwork(algorithm='random_hill_climb', max_iters=1000)
clf.fit(X_train, y_train)
test_pred = clf.predict(X_test)
train_pred = clf.predict(X_train)
print('Training')
print(classification_report(y_train, train_pred))
print('Testing')
print(classification_report(y_test, test_pred))
```

#### Tune weights with Simulated Annealing

```python
clf = mlrose.NeuralNetwork(algorithm='simulated_annealing', max_iters=1000, early_stopping=True)
clf.fit(X_train, y_train)
test_pred = clf.predict(X_test)
train_pred = clf.predict(X_train)
print('Training')
print(classification_report(y_train, train_pred))
print('Testing')
print(classification_report(y_test, test_pred))
```

#### Tune weights with Genetic Algorithm

```python
clf = mlrose.NeuralNetwork(algorithm='genetic_alg', early_stopping=True)
clf.fit(X_train, y_train)
test_pred = clf.predict(X_test)
train_pred = clf.predict(X_train)
print('Training')
print(classification_report(y_train, train_pred))
print('Testing')
print(classification_report(y_test, test_pred))
```
